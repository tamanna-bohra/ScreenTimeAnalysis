```python
max_unlocks=df['Unlocks'].max()
print(max_unlocks)

correlated_notifications=df.loc[df['Unlocks']==max_unlocks,'Notifications'].iloc[0]
print('Number of notifications the user got:',correlated_notifications)
print('No. of times user unlocked their phone:',max_unlocks)
```

```python
most_notifications = df['Notifications'].max()
corelated_unlocks = df.loc[df['Notifications'] == most_notifications, 'Unlocks'].iloc[0]
print('No. of times user unlocked their phone:', corelated_unlocks)
print('No. of max notifications:', most_notifications)
```

```python
df['Date ']= df['Date '].astype("datetime64")
```

```python
sorted_df = df.sort_values(by=['Total Usage'], ascending = False)
sorted_df.head()
```

```python
plt.figure(figsize=(15,8))
plt.style.use("dark_background")
sns.barplot(x='Date ',y='Total Usage', data = sorted_df.head(20))

plt.yticks(fontsize= 6)
plt.ylabel('Total Usage')
plt.xlabel('Date')
plt.xticks(rotation = 60)
plt.title('Total Usage per Day')
```

```python
max_opened=df['Times opened'].max()
print(max_opened)
```

```python
opened_app=df.loc[df['Times opened']==max_opened,'App'].iloc[0]
print(opened_app,"was most opened for",max_opened," times")
```

```python
max_usage = df1['Usage'].max()
_app = df1.loc[df1['Usage'] == max_usage, 'App'].iloc[0]

print(_app, 'had maximum usage =', max_usage)
```

```python
max_notification = df1['Notifications'].max()
max_app = df1.loc[df1['Notifications'] == max_notification, 'App'].iloc[0]

print(max_app, 'had max notifications =', max_notification)
```

```python
min_notification = df1['Notifications'].min()
min_app = df1.loc[df1['Notifications'] == min_notification, 'App'].iloc[0]

print(min_app, 'had least notifications =', min_notification)
```

```python
df1['Date']= df1['Date'].astype("datetime64")
```

```python
plt.figure(figsize=(15,8))
plt.style.use("dark_background")
sns.barplot(x='Date',y='Times opened', data = df1.head(20))
plt.yticks(fontsize= 6)
plt.ylabel('Times opened')
plt.xlabel('Date')
plt.xticks(rotation = 60)
plt.title('Total times apps were opened per Day')
```

```python
apps = df1['App'].value_counts()
plt.figure(figsize=(15,8))
plt.bar(apps.index, apps, color=['#FFC4C4', '#F7A76C']) 
plt.xticks(rotation = 60)
plt.show()
```
